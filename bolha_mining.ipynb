```python
from bs4 import BeautifulSoup 
import requests
import numpy as np
import pandas as pd
```

```python
def stanovanja(): 
    data = {'Cena':[], 
            'Lokacija:':[], 
            'Nadstropje:':[], 
            'Vrsta stanovanja:':[], 
            'Bivalna povr≈°ina:':[], 
            'Leto izgradnje*:':[], 
            'Leto zadnje prenove:':[]
            } 
    df = pd.DataFrame(data)
    for i in range(1,14):
        url = 'https://www.bolha.com/prodaja-stanovanja/osrednjeslovenska?page='+str(i) 
        stran = requests.get(url).text
        soup = BeautifulSoup(stran, 'lxml') 
        oglasi = soup.find_all('h3', class_ = 'entity-title') 
        for oglas in oglasi: 
            link_oglasa = 'https://www.bolha.com' + oglas.a['href'] 
            stanovanje = requests.get(link_oglasa).text
            juha = BeautifulSoup(stanovanje, 'lxml') 
            cena = juha.find('strong', class_ = 'price price--hrk').text.strip() 
            table = juha.find_all('table') 
            df2 = pd.read_html(str(table))[0].T 
            df2 = df2.rename(columns=df2.iloc[0]).drop(df2.index[0]) 
            df2.insert(0, 'Cena', [cena], True)
            df = df.append(df2, ignore_index=True) 
    return(df.to_csv('stanovanja_bolha.csv'))
stanovanja()
```
